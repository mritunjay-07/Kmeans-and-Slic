```python
from matplotlib.image import imread
import matplotlib.pyplot as plt 
```

```python
crack=imread('C:\\Users\\MRITUNJAY\\Desktop\\research paper\\DSC08006.jpg')
```

```python
plt.figure(figsize=(14,10))
plt.imshow(crack)
```

```python
crack=crack/255.0
crack.shape
```

```python
X=crack.reshape(-1,3)
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
km.fit(X)
```

```python
img_seg=km.cluster_centers_
img_seg=img_seg[km.labels_]
```

```python
img_seg=img_seg.reshape(crack.shape)
```

```python
plt.figure(figsize=(14,10))
plt.imshow(img_seg)
```

```python
km=KMeans(n_clusters=2)
km.fit(X)
img_seg=km.cluster_centers_
img_seg=img_seg[km.labels_]
img_seg=img_seg.reshape(crack.shape)
plt.figure(figsize=(14,10))
plt.imshow(img_seg)
```
